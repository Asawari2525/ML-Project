```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('/home/shubhada/Desktop/heart.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.countplot(x="target", data=df)
```

```python
y = df.target.values
```

```python
x_data = df.drop(["target"],axis=1)
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data)).values
```

```python
x_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.20,random_state=56)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
print("test accuracy {}".format(lr.score(x_test,y_test)))
lr_score=lr.score(x_test,y_test)
```

```python
y_prediction = lr.predict(x_test)
y_actual=y_test
cm = confusion_matrix(y_actual,y_prediction)
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel("Predictions Y")
plt.ylabel("Actual Y")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
print("test accuracy {}".format(knn.score(x_test,y_test)))

knn_prediction_score=knn.score(x_test,y_test)
```

```python
y_prediction = knn.predict(x_test)
y_actual=y_test
cm = confusion_matrix(y_actual,y_prediction)
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel("Predictions Y")
plt.ylabel("Actual Y")
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
svm.fit(x_train,y_train) 
print("print accuracy of svm algo: ",svm.score(x_test,y_test))

svm_score = svm.score(x_test,y_test)
```

```python
y_prediction = knn.predict(x_test)
y_actual=y_test
cm = confusion_matrix(y_actual,y_prediction)
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel("Predictions Y")
plt.ylabel("Actual Y")
plt.show()
```

```python
y_prediction = knn.predict(x_test)
y_actual=y_test
cm = confusion_matrix(y_actual,y_prediction)
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel("Predictions Y")
plt.ylabel("Actual Y")
plt.show()
```

```python
class_name = ("Logistic Regression","KNN","SVM")
class_score = (lr_score,knn_prediction_score,svm_score)
y_pos= np.arange(len(class_score))
colors = ("red","gray","purple")
plt.figure(figsize=(20,12))
plt.bar(y_pos,class_score,color=colors)
plt.xticks(y_pos,class_name,fontsize=20)
plt.yticks(np.arange(0.00, 1.05, step=0.05))
plt.ylabel('Accuracy')
plt.title(" Confusion Matrix Comparision of the Classes",fontsize=15)
plt.savefig('graph.png')
plt.show()
```
